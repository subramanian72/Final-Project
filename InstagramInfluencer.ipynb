```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
df = pd.read_csv("Influencer.csv")
```

```python
df.isnull().sum()
```

```python
df= df.dropna()
```

```python
Post=list(df['Posts'])
Post1=list()
for num in Post:
  if 'k' in num:
    new_num=1000*float(num.replace('k',''))
    Post1.append(new_num)
  else:
    Post1.append(float(num))
df['Posts']=Post1
```

```python
follows = list(df['Followers'])
follows1 = list()
for num in follows:
  if 'm' in num:
    new_num = float(num.replace('m',''))
    follows1.append(new_num)
  else:
    follows1.append(float(num))
df['Followers'] = follows1
```

```python
tlikes = list(df['Total Likes'])
tlikes1 = list()
for num in tlikes:
  if 'm' in num:
    new_num = float(num.replace('m',''))
    tlikes1.append(new_num)
  elif 'b' in num:
    new_num = float(num.replace('b',''))
    tlikes1.append(new_num)
  else:
    tlikes1.append(float(num))
df['Total Likes'] = tlikes1
```

```python
fig, ax1 = plt.subplots(figsize=(20,10))
graph = sns.countplot(ax=ax1,x = 'Country Or Region' , data = df)
graph.set_xticklabels(graph.get_xticklabels(),rotation=90)
for p in graph.patches:
    height = p.get_height()
    graph.text(p.get_x()+p.get_width()/2., height + 0.1,height ,ha="center")
```

```python
df.columns
```

```python
for column_name in df.columns:
    unique_values = len(df[column_name].unique())
    print("Feature '{column_name}' has '{unique_values}' unique values".format(column_name = column_name, unique_values=unique_values))
```

```python
num_features=['Influence Score' ,'Followers','Posts']
```

```python
for column_name in num_features:
   plt.figure(figsize=(7, 7))
   plt.title("{column_name} Distribution".format(column_name = column_name))
   sns.histplot(df[column_name], stat='density', color='#E6E6FA')
   sns.kdeplot(df[column_name], color='black')
   plt.axvline(df[column_name].mean(), color='red', linestyle='--', linewidth=0.8)
   min_ylim, max_ylim = plt.ylim()
   plt.text(df[column_name].mean()*1.05, max_ylim*0.96, 'Mean (μ): {:.2f}'.format(df[column_name].mean()))
   plt.xlabel("Score")
   plt.ylabel("Density")
   plt.show()
```

```python
for column in num_features:
  top10 = df.nlargest(10,column)
  plt.figure(figsize=(20,8))
  plt.bar(top10['Channel Info'], top10[column],color='#D8BFD8')
  plt.ylabel(column, fontsize = 20)
  plt.xlabel("User name", fontsize = 20)
  plt.title("Top 10 blogers per y", fontsize = 20)
```

```python
sns.set()
sns.pairplot(df,height = 5 ,kind ='reg',diag_kind='kde')
plt.show()
```

```python
i=1
plt.figure(figsize=(15,25))
for feature in num_features:
    plt.subplot(6,3,i)
    sns.boxplot(y=df[feature],color='#DDA0DD')
    i+=1
```

```python
correlation = df.corr()
print(correlation['Total Likes'].sort_values(ascending = False),'\n')
```
